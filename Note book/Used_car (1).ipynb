```python
df['posting_date'] = pd.to_datetime(df['posting_date'], utc=True)

df['posting_month'] = df['posting_date'].dt.month
df['posting_day'] = df['posting_date'].dt.day
df['posting_weekday'] = df['posting_date'].dt.weekday 
df['posting_day'] = df['posting_date'].dt.day_name()
```

```python
df['has_phone'] = df['description'].str.contains(r'\d{3}-\d{3}-\d{4}', regex=True).astype(int)
df['desc_word_count'] = df['description'].apply(lambda x: len(str(x).split()))
df['has_warranty'] = df['description'].str.contains('warranty', case=False).astype(int)
df['is_one_owner'] = df['description'].str.contains('one owner|1st owner', case=False).astype(int)
```
